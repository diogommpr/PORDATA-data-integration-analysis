```python
# =========================================
# Import required libraries
# =========================================

import re
import pandas as pd
import numpy as np
import unidecode # normalize strings by removing accents
```

```python
# =========================================
# Load CSV datasets into separate DataFrames
# =========================================

raw_data_path = "../data/raw"

# Demography
population_df = pd.read_csv(f"{raw_data_path}\population_by_age_sex.csv")

# Mortality 
mortality_df = pd.read_csv(f"{raw_data_path}\mortality_rate.csv")
infant_mortality_df = pd.read_csv(f"{raw_data_path}\infant_mortality_rate.csv")

# Natality 
natality_df = pd.read_csv(f"{raw_data_path}\birth_rate.csv")

# Unemployment
unemployment_df = pd.read_csv(f"{raw_data_path}\unemployment_by_age.csv")

# Income
income_df = pd.read_csv(f"{raw_data_path}\average_monthly_income.csv")

# Pharmacies
pharmacies_df = pd.read_csv(f"{raw_data_path}\pharmacies.csv")

# Crime
crimes_df = pd.read_csv(f"{raw_data_path}\crimes_by_category.csv")

# Marital indicators 
nuptiality_df = pd.read_csv(f"{raw_data_path}\marriage_rate.csv")
divorce_df = pd.read_csv(f"{raw_data_path}\divorce_rate.csv")
```

```python
# =========================================
# Quick inspection of all datasets
# =========================================

# Create dictionary to hold dataset names and DataFrames
datasets = {
    "População Residente": population_df,
    "Taxa Bruta de Mortalidade": mortality_df,
    "Taxa de Mortalidade Infantil": infant_mortality_df,
    "Taxa Bruta de Natalidade": natality_df,
    "Taxa de Desemprego": unemployment_df,
    "Salário": income_df,
    "Farmácias": pharmacies_df,
    "Crimes": crimes_df,
    "Taxa de Nupcialidade": nuptiality_df,
    "Taxa de Divórcio": divorce_df
}

# Iterate through datasets for a quick inspection
for name, df in datasets.items():
    print(f"\n{name}:")
    display(df.head())
    print("\nInfo:")
    print(df.info())
    print("\nSummary statistics:")
    print(df.describe())
```

```python
# =========================================
# Clean column names for all datasets
# =========================================

for df in datasets.values():
    # Remove extra spaces (leading and trailing)
    df.columns = df.columns.str.strip() 

    # Replace spaces with underscores
    df.columns = df.columns.str.replace(" ", "_", regex=False) 

    # Remove accents for consistency 
    df.columns = [unidecode.unidecode(col) for col in df.columns]

    # Remove starting numbers for consistency (they vary across datasets)
    df.columns = df.columns.str.replace(r'^\d+\._', "", regex=True) 

# Print cleaned columns names to confirm
for name, df in datasets.items():
    print(f"{name} columns:", df.columns.tolist())
```

```python
# =========================================
# Filter datasets to ensure one value per Municipio
# =========================================

# Rationale:
# Some datasets contain multiple breakdowns (by age, sex, etc.).

# Population dataset: keep only totals 
datasets["População Residente"] = datasets["População Residente"][
    (datasets["População Residente"]["Filtro_1"] == "Total") &
    (datasets["População Residente"]["Filtro_2"] == "Total")
]

# Unemployment dataset: keep only totals 
datasets["Taxa de Desemprego"] = datasets["Taxa de Desemprego"][
    datasets["Taxa de Desemprego"]["Filtro_1"] == "Total"
]

# Crimes dataset: keep only totals
datasets["Crimes_Total"] = datasets["Crimes"][
    datasets["Crimes"]["Filtro_1"] == "Total"
]
```

```python
# =========================================
# Remove unnecessary columns from all datasets
# =========================================

# Columns to drop (not needed for analysis)
cols_to_drop = [
    "Nome_Pais_(Europa)", 
    "Filtro_1", 
    "Filtro_2", 
    "Filtro_3", 
    "Escala", 
    "Simbolo",
]

for name, df in datasets.items():
    # Keep Filtro_1 (type of crime) for crimes_df, drop everything else
    if name == "Crimes":
        df.drop(columns=[col for col in cols_to_drop if col != "Filtro_1"], inplace=True, errors="ignore")
    else:
        df.drop(columns=cols_to_drop, inplace=True, errors="ignore") # ignores missing columns

# Print cleaned column names to confirm
for name, df in datasets.items():
    print(f"{name} columns:", df.columns.tolist())
```

```python
# =========================================
# Rename column names 
# =========================================

# New column names
new_names = {
    "Nome_Regiao_(Portugal)": "Regiao",
    "Ambito_Geografico": "Nivel",
}

# Map each dataset's "Valor" column to a descriptive name 
valor_names = {
    "População Residente": "Populacao",
    "Taxa Bruta de Mortalidade": "Mortes_por_1000",
    "Taxa de Mortalidade Infantil": "Mortes_por_1000_nadosvivos",
    "Taxa Bruta de Natalidade": "Nascimentos_por_1000",
    "Taxa de Desemprego": "n_desempregados",
    "Salário": "Salario",
    "Farmácias": "n_farmacias",
    "Crimes_Total": "n_crimes",
    "Crimes": "n_crimes",
    "Taxa de Nupcialidade": "Casamentos_por_1000",
    "Taxa de Divórcio": "Divorcios_por_1000",
}

# Rename columns
for name, df in datasets.items():
    # Rename column names
    df.rename(columns=new_names, inplace=True) 

    # Rename "Valor" column to dataset-specific descriptive name
    if name in valor_names:
        df.rename(columns={"Valor": valor_names[name]}, inplace=True) 

# Print cleaned column names to confirm
for name, df in datasets.items():
    print(f"{name} columns:", df.columns.tolist())
```

```python
# =========================================
# Standardize values for Regiao and Nivel 
# =========================================

# Standardize Regiao and Nivel columns in all datasets
for df in datasets.values():
    # Remove leading/trailing spaces, capitalize, remove accents
    df["Regiao"] = df["Regiao"].astype(str).str.strip().str.title() 
    df["Regiao"] = df["Regiao"].apply(lambda x: unidecode.unidecode(x)) 
    
    df["Nivel"] = df["Nivel"].astype(str).str.strip().str.title() 
    df["Nivel"] = df["Nivel"].apply(lambda x: unidecode.unidecode(x)) 

# Special case: Crimes dataset
crimes = datasets["Crimes"]

# Rename column Filro_1 to Tipo_Crime
crimes.rename(columns={"Filtro_1": "Tipo_Crime"}, inplace=True)

# Standardize Tipo_Crime values
crimes["Tipo_Crime"] = (
    crimes["Tipo_Crime"]
    .astype(str)
    .str.strip()
    .str.title() # removes leading/trailing spaces and capitalizes each word
    .apply(lambda x: re.sub(r'^\d+\.\s*', '', x))  # Remove "01. ", "02. ", etc.
    .apply(lambda x: unidecode.unidecode(x)) # remove accents
)

# Check unique Regiao values for all datasets

for name, df in datasets.items():
    print(f"{name} unique Regiao values:", df["Regiao"].unique())
```

```python
# =========================================
# Filter datasets to include only "Municipio" level 
# =========================================

# Function to filter DataFrame for Nivel = "Municipio"
def filter_municipio(df):
    # Keeps only rows where Nivel column = "Municipio".
    # Returns a copy of the filtered dataframe. 
    return df[df["Nivel"] == "Municipio"].copy()

# Apply filter to all datasets
for name, df in datasets.items():
    datasets[name] = filter_municipio(df)

# Check filtered dataset sample
print(datasets["População Residente"].head())
```

```python
# =========================================
# Drop "Nivel" column from all datasets
# =========================================

for df in datasets.values():
    df.drop(columns=["Nivel"], inplace=True, errors="ignore")

# Check a dataset
print(datasets["População Residente"].head())
```

```python
# =========================================
# Carry over values from 2023 to 2024 for selected datasets
# =========================================

# Rationale:
# Some datasets do not yet have values for 2024

carry_over_years = ["Salário", "Taxa de Divórcio"] # keys from datasets dictionary

for name in carry_over_years:
    df = datasets[name]

    # Select rows from 2023
    df_2023 = df[df["Ano"] == 2023].copy()

    # Update year to 2024
    df_2023["Ano"] = 2024

    # Append to original dataset
    datasets[name] = pd.concat([df, df_2023], ignore_index=True)
```

```python
# =========================================
# Merge all datasets into a single consolidated dataframe
# =========================================

# Merge keys: Municipality (Regiao) and Year (Ano)
merge_keys = ["Ano", "Regiao"]

# Start with population dataset
merged_df = datasets["População Residente"]

# List of datasets to merge (all except population)
merge_order = [
    "Taxa Bruta de Mortalidade",
    "Taxa de Mortalidade Infantil",
    "Taxa Bruta de Natalidade",
    "Taxa de Desemprego",
    "Salário",
    "Farmácias",
    "Crimes_Total",
    "Taxa de Nupcialidade",
    "Taxa de Divórcio"
]

# Merge each dataset on the same keys
for key in merge_order:
    df_to_merge = datasets[key]
    merged_df = merged_df.merge(df_to_merge, on=merge_keys, how="left")

# Check for missing values after merging
print(merged_df.isna().sum())

# Sort merged dataset by Regiao and Ano and reset index
merged_df.sort_values(by=["Regiao", "Ano"], inplace=True)
merged_df.reset_index(drop=True, inplace=True)
```

```python
# =========================================
# Merge all datasets into a single detailed dataframe (keeps Tipo_Crime)
# =========================================

# Merge keys: Municipality (Regiao) and Year (Ano)
merge_keys = ["Ano", "Regiao"]

# Start with population dataset
merged_detailed_df = datasets["População Residente"]

# List of datasets to merge (all except population)
merge_order_detailed = [
    "Taxa Bruta de Mortalidade",
    "Taxa de Mortalidade Infantil",
    "Taxa Bruta de Natalidade",
    "Taxa de Desemprego",
    "Salário",
    "Farmácias",
    "Crimes", # keeps multiple rows per Tipo_Crime
    "Taxa de Nupcialidade",
    "Taxa de Divórcio"
]

# Merge each dataset on the same keys
for key in merge_order_detailed:
    df_to_merge = datasets[key]
    merged_detailed_df = merged_detailed_df.merge(df_to_merge, on=merge_keys, how="left")

# Check for missing values after merging
print(merged_detailed_df.isna().sum())

# Sort merged dataset by Regiao and Ano and reset index
merged_detailed_df.sort_values(by=["Regiao", "Ano"], inplace=True)
merged_detailed_df.reset_index(drop=True, inplace=True)
```

```python
# =========================================
# Save cleaned and merged datasets to CSV
# =========================================

processed_data_path = "../data/processed"

# Export standard merged dataset
merged_df.to_csv(f"{processed_data_path}\merged_dataset.csv", index=False)
print("CSV saved successfully. Shape:", merged_df.shape)

# Export detailed merged dataset (keeps multiple row per Tipo_Crime)
merged_detailed_df.to_csv(f"{processed_data_path}\merged_dataset_detailed.csv", index=False)
print("CSV saved successfully. Shape:", merged_detailed_df.shape)
```

```python
# =========================================
# Analyse year 2024
# =========================================

def inspect_2024(df, name):
    # Filter for 2024
    df_2024 = df[df["Ano"] == 2024].copy()

    # Count missing "Regiao"
    na_count = df_2024["Regiao"].isna().sum()

    # Count rows corresponding to Portugal (should be 0)
    portugal_count = (df_2024["Regiao"] == "Portugal").sum()

    # Print results
    print(f"{name} - Year 2024:")
    print(f"Missing 'Regiao' in 2024: {na_count}")
    print(f"Rows for 'Portugal' in 2024: {portugal_count}")
    
    return df_2024

# Apply to merged_df
df_2024_simple = inspect_2024(merged_df, "Merged (one row per municipality)")

# Apply to merged_detailed_df
df_2024_detailed = inspect_2024(merged_detailed_df, "Merged Detailed (multiple rows per municipality)")
```

```python
# =========================================
# Save 2024-specific datasets to CSV
# =========================================

# Save simple 2024 dataset (one row per municipality)
df_2024_simple.to_csv(f"{processed_data_path}\municipal_data_2024.csv", index=False)
print("Simple 2024 CSV saved. Shape:", df_2024_simple.shape)

# Save detailed 2024 dataset (multiple rows per Tipo_Crime)
df_2024_detailed.to_csv(f"{processed_data_path}\municipal_data_2024_detailed.csv', index=False)
print("Detailed 2024 CSV saved. Shape:", df_2024_detailed.shape)
```
